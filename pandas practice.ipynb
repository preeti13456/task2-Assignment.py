```python
#comparing two csv files using pandas
A = pd.read_csv("data.csv", header=None, usecols=[0], names=['col']).drop_duplicates()
B = pd.read_csv("autos.csv", header=None, usecols=[0], names=['col']).drop_duplicates()
# A - B
```

```python
pd.merge(A, B, on='col', how='left', indicator=True).query("_merge == 'left_only'")
```

```python
# B - A
pd.merge(A, B, on='col', how='right', indicator=True).query("_merge == 'right_only'")
```
