```python
import numpy as np
t=np.arange(1,20,0.5)
plt.plot(t,t**2,'b--',label='^2')
plt.plot(t,t**2.2,'rs',label='^2.2')
plt.plot(t,t**2.5,'g^',label='^2.5')
plt.xlabel("Numpy array")
plt.ylabel("Calculations")
plt.title("MyPlot")
plt.legend()
plt.grid()
plt.show()
```
